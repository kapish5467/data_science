```python
t["Age"].fillna(t.Age.mean(), inplace=True)
t.describe()
```

```python
y=t.pop("Survived")
```

```python
def describe_categorical(t):
    """
    Just like .describe(), but returns the results for
    categorical variables only.
    """
    from IPython.display import display, HTML
    display(HTML(t[t.columns[t.dtypes == "object"]].describe().to_html()))
```

```python
describe_categorical(t)
```

```python
t.drop(["Name","Ticket","PassengerId"], axis=1, inplace=True)
```

```python
#Changing the cabin variables to only first letter or none
def new_cabin(x):
    try:
        return x[0]
    except TypeError:
        return "none"
t["Cabin"] = t.Cabin.apply(new_cabin)
```

```python
categorical_var = ['Sex', 'Cabin', 'Embarked']

for variable in categorical_var:
    t[variable].fillna("Missing", inplace=True)
    dummies = ds.get_dummies(t[variable], prefix=variable)
    t = ds.concat([t, dummies], axis=1)
    t.drop([variable], axis=1, inplace=True)
```
