```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
import math

%pylab inline
```

```python
data = pd.read_csv("breast_cancer.csv")
```

```python
data.head()
```

```python
data[data.malignant==1]
```

```python
data.columns
```

Columns "id number" and "Unnamed" do not contribute to our model and hence can be dropped.

```python
data = data.drop(['Unnamed: 0','id number'], axis = 1)
```

```python
data.columns
```

```python
y = data.pop("malignant")
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=.2, random_state=42)
```

Grid Search

```python
n_estimators = [300,400,500]
max_features = ['auto', 'sqrt', 'log2']
min_samples_split = [1,2,3,4]
rfc = RandomForestClassifier(n_jobs=1)
estimator = GridSearchCV(rfc,
                        dict(n_estimators=n_estimators,
                            max_features=max_features,
                            min_samples_split=min_samples_split
                            ), cv=None, n_jobs=-1)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
best_rfc = estimator.best_estimator_
```

```python
accuracy = accuracy_score(y_test, best_rfc.predict(X_test))
print "Accuracy: ", accuracy
```

```python
y_hat = best_rfc.predict(X_test)
```

```python
correct = 0
total = y_test.shape[0]
for pred_val, truth_val in zip(y_hat, y_test):
    if pred_val == truth_val:
        correct+=1
        
print correct / float(total)
```

Precision is called positive predictive value which specifies the percentage when the patients have malignant breast cancer.

```python
print classification_report(y_test, best_rfc.predict(X_test))
```

```python
print confusion_matrix(y_test, best_rfc.predict(X_test))
```

Precission and Recall

<li> Number of false positives for the number of patients have breast cancer which is malignant(fp) = 2
<li> Number of true negatives for the number of patients have breast cancer which is malignant(fn)= 3
<li> Precision = tp/tp+fp
<li> Precision: 42 / 42 +  2 = .954545
<li> Recall = tp/tp+fn
<li> Recall: 42 / 42 + 3 = .93333
