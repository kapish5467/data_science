Lets analyse which age groups of the passengers survived the disaster.

```python
age_group1= df[df.Age<=15].Survived.value_counts().plot(kind='bar')
age_group1.set_title("Age Group of Passengers between 0 to 15")
age_group1.set_xlabel("Survived")
age_group1.set_ylabel("Count of Passengers")
```

```python
age_group1= df[(df.Age>15) & (df.Age<30) ].Survived.value_counts().plot(kind='bar')
age_group1.set_title("Age Group of Passengers between 15 to 30")
age_group1.set_xlabel("Survived")
age_group1.set_ylabel("Count of Passengers")
```

```python
age_group3= df[(df.Age>30) & (df.Age<=60) ].Survived.value_counts().plot(kind='bar')
age_group3.set_title("Age Group of Passengers between 30 to 60")
age_group3.set_xlabel("Survived")
age_group3.set_ylabel("Count of Passengers")
```

```python
age_group4= df[df.Age>60].Survived.value_counts().plot(kind='bar')
age_group4.set_title("Age Group of Passengers more than 60")
age_group4.set_xlabel("Survived")
age_group4.set_ylabel("Count of Passengers")
```
