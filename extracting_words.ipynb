```python
import os

path = "C:/Users/Karan/Documents/data_science/GRS/newcorpus/sethi"
def get_tokens():
    tokens = []
    for filename in os.listdir(path):
        filepath = path+"/"+filename
        _f  = open(filepath,'r')
        text = _f.read()
        lowers = text.lower()
        no_punctuation = lowers.translate(None, string.punctuation)
        file_tokens = nltk.word_tokenize(no_punctuation)
        for word in file_tokens:
            tokens.append(word)
    return tokens


tokens = get_tokens()
```

```python
from nltk.corpus import stopwords

filtered_all = [w for w in tokens if not w in stopwords.words('english')]
print filtered_all
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
def lemma_tokens(tokens):
    lemmatized = []
    for item in tokens:
        try:
            lemmatized.append(wordnet_lemmatizer.lemmatize(item))
        except UnicodeDecodeError:
            continue
    return lemmatized

lemmatized = lemma_tokens(filtered_all)
_f = open("C:/Users/Karan/Documents/data_science/GRS/all_files_lemmas.txt","w")
for words in lemmatized:
    _f.write(str((words.decode("utf-8"))))
    _f.write('\n')

_f.close()
```
