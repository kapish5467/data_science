```python
#categorical_variables = ['holiday','season','weather','workingday']

cat_variables = ['holiday','season','weather','workingday']
for variable in cat_variables:
    dummies = pd.get_dummies(training[variable], prefix = variable)
    training = pd.concat([training, dummies], axis=1)
    training.drop([variable], axis=1, inplace=True)
```

```python
training.columns
```

Now because the prediction has to be made by kayak rented during each hour, the value of hour has to be extracted from the datetime attribute to correctly predict the data.

```python
datetime_extracted_train = pd.DatetimeIndex(training['datetime'])
training['year'] = datetime_extracted_train.year
training['month'] = datetime_extracted_train.month
training['weekday'] = datetime_extracted_train.weekday
training['hour'] = datetime_extracted_train.hour
```

```python
datetime_extracted_test = pd.DatetimeIndex(testing['datetime'])
testing['year'] = datetime_extracted_test.year
testing['month'] = datetime_extracted_test.month
testing['weekday'] = datetime_extracted_test.weekday
testing['hour'] = datetime_extracted_test.hour
```

```python
all_columns = ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered','holiday_0', 'holiday_1', 'season_1', 'season_2', 'season_3', 'season_4', 'weather_1', 'weather_2','weather_3','weather_4', 'workingday_0', 'workingday_1', 'year', 'month', 'weekday', 'hour']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.grid_search import GridSearchCV
from datetime import datetime
temp = pd.DatetimeIndex(training['datetime'])
train = training[temp.day <= 17]
validation = training[temp.day > 17]
 
param_grid = {'min_samples_leaf': [15]}
 
model = RandomForestRegressor(n_estimators=2000, oob_score=True)

gs_cv = GridSearchCV(
    model, param_grid, n_jobs=-1, cv=10).fit(
    train[all_columns], train['count'])
 
gs_cv.best_params_
error_count = mean_squared_error(validation['count'], gs_cv.predict(validation[all_columns]))
 
result = gs_cv.predict(testing[all_columns])
df=pd.DataFrame({'datetime':testing['datetime'], 'count':result})
df.to_csv('RF_results.csv', index = False, columns=['datetime','count'])
```
